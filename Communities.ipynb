```python
# importing the used libs
import random
import networkx as netx
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from networkx.algorithms import community
```

```python
G = netx.Graph()
# reading the graph file
graph = f = open("netM.csv", "r")

# understanding the data
lines = graph.readlines()

for line in lines:
  no_whitespace = line.strip()
  nodes = no_whitespace.split(" ")
  G.add_nodes_from(nodes[0])
  for i, node in enumerate(nodes):
    if i != 0:
      G.add_edge(nodes[0], node)

print(len(G.nodes))
print(len(G.edges))
```

```python
# Using the function asyn_fluidc
comm = community.asyn_fluid.asyn_fluidc(G, 5)
asyn_fluid_result = list(comm)
asyn_fluid_result.sort(key=len)
print(len(asyn_fluid_result))
it = 1

# writing results to txt file
with open('asyn_fluidc.txt', 'w+') as f:
  for result_fluid in asyn_fluid_result:
    f.write("Community " + str(it) + " of " + str(len(asyn_fluid_result)))
    print(sorted(result_fluid))
    line = sorted(result_fluid)
    f.write('%s \n' %line)
    it+=1
f.close()
```

```python
#using the function asyn_lpa_communities
it = 1
comm2 = community.asyn_lpa_communities(G)
asyn_lpa = list(comm2)
asyn_lpa.sort(key=len)
print(len(asyn_lpa))

# writing results to txt file
with open('asyn_lpa.txt', 'w+') as f:
  for result_lpa in asyn_lpa:
    f.write("Community " + str(it) + " of " + str(len(asyn_lpa)))
    print(sorted(result_lpa))
    line = sorted(result_lpa)
    f.write('%s \n' %line)
    it +=1
f.close()
```

The results of number of communities are equal using both functions and visually. The members of each community tend to be equal either.
